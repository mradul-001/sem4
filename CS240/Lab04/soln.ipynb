```python
import numpy as np
import matplotlib.pyplot as plt
```

### Input for XOR

```python
x_xor = np.array([[0, 0],[0, 1],[1, 0],[1, 1]])
y_xor = np.array([0, 1, 1, 0])
```

### Input for Palindrome

```python
palindrome_4 = []
for i in [0, 1]:
    for j in [0, 1]:
        for k in [0, 1]:
            for l in [0, 1]:
                palindrome_4.append([i, j, k, l])
```

```python
y_palindrome_4 = []
for ele in palindrome_4:
    if ele == ele[::-1]:
        y_palindrome_4.append(1)
    else:
        y_palindrome_4.append(0)
```

```python
y_palindrome_4 = np.array(y_palindrome_4)
palindrome_4 = np.array(palindrome_4)
```

```python
palindrome_5 = []
for i in [0, 1]:
    for j in [0, 1]:
        for k in [0, 1]:
            for l in [0, 1]:
                for m in [0, 1]:
                    palindrome_5.append([i, j, k, l, m])
```

```python
y_palindrome_5 = []
for ele in palindrome_5:
    if ele == ele[::-1]:
        y_palindrome_5.append(1)
    else:
        y_palindrome_5.append(0)
```

```python
y_palindrome_5 = np.array(y_palindrome_5)
palindrome_5 = np.array(palindrome_5)
```

### Input for majority function

```python
majority_4 = palindrome_4
majority_5 = palindrome_5
```

```python
y_majority_4 = []
for ele in majority_4:
    count = 0
    for i in ele:
        if i == 1:
            count += 1
    if count > 2:
        y_majority_4.append(1)
    else:
        y_majority_4.append(0)
```

```python
y_majority_5 = []
for ele in majority_5:
    count = 0
    for i in ele:
        if i == 1:
            count += 1
    if count > 2:
        y_majority_5.append(1)
    else:
        y_majority_5.append(0)
```

```python
y_majority_4 = np.array(y_majority_4)
y_majority_5 = np.array(y_majority_5)
```

### Input for even parity

```python
eParity_4 = palindrome_4
eParity_5 = palindrome_5
```

```python
y_eParity_4 = []
for ele in eParity_4:
    count = 0
    for i in ele:
        if i == 1:
            count += 1
    if count % 2 == 0:
        y_eParity_4.append(1)
    else:
        y_eParity_4.append(0)
```

```python
y_eParity_5 = []
for ele in eParity_5:
    count = 0
    for i in ele:
        if i == 1:
            count += 1
    if count % 2 == 0:
        y_eParity_5.append(1)
    else:
        y_eParity_5.append(0)
```

```python
y_eParity_4 = np.array(y_eParity_4)
y_eParity_5 = np.array(y_eParity_5)
```

## Model architecture and backward propogation

```python
# function to initialize the weights
def arch(shape):
    w = np.random.random(shape)
    b = np.random.random((1, 1))[0][0]
    return w, b
```

```python
# sigmoid activation
def sigmoid(a):
    return 1 / (1 + np.exp(-a))
```

```python
# function to update weights
def updateWeights(w, b, x, y, alpha):

    y = y.reshape(-1, 1)
    z = np.dot(x, w) + b
    a = sigmoid(z)
    
    error = 1 / 2 * np.sum((y - a)**2)
    
    i = np.multiply(y - a, a)
    j = (1 - a)
    
    dw =  np.dot(x.T, np.multiply(i, j)) * alpha
    db =  np.sum(np.multiply(i, j)) * alpha

    k = ((y - a) * a * (1 - a))
    
    w = w + dw
    b = b + db

    return w, b, error
```

### Palindrome 4

```python
w, b = arch((4, 1))
err = []
for epoch in range(10000):
    w, b, error = updateWeights(w, b, palindrome_4, y_palindrome_4, 0.00050)
    err.append(error)
```

```python
plt.figure(figsize = (15, 8))
plt.title("Loss vs. Epochs")
plt.xlabel("No. of iterations")
plt.ylabel("Mean Squared Error")
plt.plot(err)
```
